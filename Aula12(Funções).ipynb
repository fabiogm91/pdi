```python
import tifffile as tiff
from spectral import imshow
import numpy as np
```

```python
img = tiff.imread('/content/L71221071_07120010720_DN.tif')
```

```python
azul = img[:,:,0]
verde = img[:,:,1]
vermelho = img[:,:,2]
```

```python
imshow(vermelho)
```

```python
media = lambda bx,by,bz: (bx + by + bz) /3
```

```python
m= media(azul, verde, vermelho)
```

```python
imshow(m, figsize=(12,8))
```
